```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
req = requests.get("https://biopharmguy.com/links/company-by-name-southern-california.php")
```

```python
page = req.text
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
company_location = soup.find_all("td", "location")
location = [x.text for x in company_location]
company_description = soup.find_all("td", "description")
description = [x.text for x in company_description]
company_name = soup.find_all("td", "company")
company = [x.text for x in company_name]
```

```python
zipped_list = list(zip(company, location, description))
df = pd.DataFrame(zipped_list, columns = ["Company", "Location", "Description"])
zipped_list
```
